```python
#importing required libraries
from skimage.measure import compare_ssim
import imutils
import cv2 

import matplotlib.pyplot as plt
```

```python
#loading the images into the notebook
from google.colab import files
files.upload()
```

```python
#storing the images 
img1 = cv2.imread("art.png")
img2 = cv2.imread("art2.png")
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
#calculating the structural similarity index and getting the diff image 
(score, diff) = compare_ssim(img1, img2, full=True, multichannel=True)
diff = (diff*255).astype('uint8')
print(' SSIM score = {}'.format(score))
```

Similarity score of 0.968 suggests that image 2 is a copy of image 1 with a slight change.

```python
#plotting the difference image 
plt.imshow(diff)
```

The difference image shows that change that was found in image 2. Other than that, the rest of the image is same.

#Creating bbox

```python
grayimg = cv2.cvtColor(diff, cv2.COLOR_BGR2GRAY)
```

```python
#thresholding the difference image
thresh = cv2.threshold(grayimg, 0, 255, cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)[1]
#getting the contours
contr = cv2.findContours(thresh.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
contr = imutils.grab_contours(contr)
```

```python
#looping through the contours
for c in contr:
  x, y, width, height = cv2.boundingRect(c)
  cv2.rectangle(img1, (x, y), (x + width, y + height), (255, 0, 0), 2)
  cv2.rectangle(img2, (x, y), (x + width, y + height), (255, 0, 0), 2)
```

```python
#calculating the structural similarity index and getting the diff image 
(score_2, diff_2) = compare_ssim(img_1, img_2, full=True, multichannel=True)
diff_2 = (diff_2*255).astype('uint8')
print(' SSIM score = {}'.format(score_2))
```

#Creating bboxes on second image

```python
gimg = cv2.cvtColor(diff_2, cv2.COLOR_BGR2GRAY)

#thresholding the difference image
thresh_2 = cv2.threshold(gimg, 0, 255, cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)[1]
#getting the contours
contr_2 = cv2.findContours(thresh_2.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
contr_2 = imutils.grab_contours(contr_2)

#looping through the contours
for c in contr_2:
  x, y, width, height = cv2.boundingRect(c)
  cv2.rectangle(img_1, (x, y), (x + width, y + height), (255, 0, 0), 2)
  cv2.rectangle(img_2, (x, y), (x + width, y + height), (255, 0, 0), 2)
```
